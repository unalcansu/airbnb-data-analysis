```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import statsmodels.api as sm
```

```python
df = pd.read_csv('Bologna/airbnb_bologna_0310_2016-02-07.csv')
```

```python
df.drop(['room_id', 'survey_id', 'host_id', 'country', 'city', 'neighborhood', 'borough', 'name', 'last_modified', 'latitude', 'longitude', 'location'], axis=1, inplace=True)
df.dropna(inplace=True)
df
```

```python
variables = ['reviews', 'overall_satisfaction', 'accommodates', 'bedrooms', 'bathrooms', 'price', 'minstay' ]
```

```python
df['overall_satisfaction'].fillna(df['overall_satisfaction'].mode()[0], inplace=True)
df['accommodates'].fillna(df['accommodates'].mode()[0], inplace=True)
df['bedrooms'].fillna(df['bedrooms'].mode()[0], inplace=True)
df['bathrooms'].fillna(df['bathrooms'].mode()[0], inplace=True)
df['minstay'].fillna(df['minstay'].mode()[0], inplace=True)
```

```python
df = pd.get_dummies(df, columns=['room_type'])
df.head()
```

```python
#Entire home/apt model
f = df.loc[df['room_type_Entire home/apt'] == 1, :]
f_data = f.copy()
f_data = f.dropna(subset=variables)
model = sm.OLS(f_data['overall_satisfaction'], sm.add_constant(f_data.loc[:, variables]))
f_results = model.fit()
f_results.summary()
```

```python
#Private room model
f = df.loc[df['room_type_Private room'] == 1, :]
f_data = f.copy()
f_data = f.dropna(subset=variables)
model = sm.OLS(f_data['overall_satisfaction'], sm.add_constant(f_data.loc[:, variables]))
f_results = model.fit()
f_results.summary()
```

```python
#Shared room model
f = df.loc[df['room_type_Shared room'] == 1, :]
f_data = f.copy()
f_data = f.dropna(subset=variables)
model = sm.OLS(f_data['overall_satisfaction'], sm.add_constant(f_data.loc[:, variables]))
f_results = model.fit()
f_results.summary()
```
